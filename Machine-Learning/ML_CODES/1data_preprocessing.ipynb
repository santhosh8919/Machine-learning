```python
from sklearn.impute import SimpleImputer
import numpy as np

data = np.array([[1, 2, np.nan],  
                 [3, 4, 5],   
                 [6, np.nan, 8], 
                 [9, 10, 11]])     
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputed_data = imputer.fit_transform(data)
print(imputed_data)  


```
