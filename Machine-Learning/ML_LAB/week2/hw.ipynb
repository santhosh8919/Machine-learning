```python
#12: Write a function to find min and max values of any column
def find_min_max(df, column_name):
    if column_name in df.columns:
        min_value = df[column_name].min()
        max_value = df[column_name].max()
        
        # Find rows with min and max values
        min_row = df[df[column_name] == min_value]
        max_row = df[df[column_name] == max_value]
        
        return min_value, max_value, min_row, max_row
    else:
        return None, f"Error: Column '{column_name}' not found in DataFrame."

# Example usage
column = "Horsepower"
min_val, max_val, min_row, max_row = find_min_max(df, column)
print(f"\nMin {column}:", min_val, "\n", min_row)
print(f"\nMax {column}:", max_val, "\n", max_row)
```

```python
#Q13: Call the above function for horsepower, length, fuel efficiency
for col in ["Horsepower", "Length", "Fuel efficiency"]:
    min_val, max_val, min_row, max_row = find_min_max(df, col)
    print(f"\n{col} - Min: {min_val} \n", min_row)
    print(f"\n{col} - Max: {max_val} \n", max_row)
```

```python
#23: Divide data into X (inputs) & y (output)
y=df['Sales in thousands']
x=df.drop(columns=['Sales in thousands'])
print("x shape:",x.shape)
print("y shape",y.shape)
```

```python
#24 Encode categorical variables using Label Encoding
from sklearn.preprocessing import LabelEncoder

categorical_columns = x.select_dtypes(include=['object']).columns
label_encoders = {}

for col in categorical_columns:
    le = LabelEncoder()
    x[col] = le.fit_transform(x[col])
    label_encoders[col] = le

print(x.head())
```

```python
#25 : Encode "Vehicle Type" using One-Hot Encoding
x = pd.get_dummies(x, columns=['Vehicle type'], prefix='Vehicle', drop_first=True)
print(x.head())
```

```python
# 26: Split data into train (70%) & test (10%)
from sklearn.model_selection import train_test_split

X_train, X_temp, y_train, y_temp = train_test_split(x, y, test_size=0.30, random_state=42)
X_test, _, y_test, _ = train_test_split(X_temp, y_temp, test_size=2/3, random_state=42)

print("Training set:", X_train.shape, y_train.shape)
print("Test set:", X_test.shape, y_test.shape)

```
