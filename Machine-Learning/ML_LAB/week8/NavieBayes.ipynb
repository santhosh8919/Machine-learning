```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
newsgroups_train=pd.read_csv('newsgroups_train.csv')
```

```python
newsgroups_train.shape
```

```python
newsgroups_train.iloc[0,:]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vec=CountVectorizer(max_features=1500)
vectors=count_vec.fit_transform(newsgroups_train.text)
vectors=vectors.toarray()
vectors
```

```python
X=vectors
y=newsgroups_train.target
```

```python
from sklearn.naive_bayes import BernoulliNB
clf=BernoulliNB(alpha=.01)
clf.fit(X,y)
```

```python
from sklearn.metrics import accuracy_score

y_pred = clf.predict(X)
accuracy = accuracy_score(y, y_pred)

print("Accuracy Score:", accuracy)
```

```python
test_data = pd.read_csv('newsgroups_test.csv')
```

```python
from sklearn.metrics import accuracy_score, classification_report

X_test = count_vec.transform(test_data['text']).toarray()
y_test = test_data['target']

y_pred = clf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)


accuracy
```
