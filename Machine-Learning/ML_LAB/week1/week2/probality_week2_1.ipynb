```python
import pandas as pd
from scipy.stats import gaussian_kde

#load dataset
df=pd.read_csv("Car_sales.csv")
df.columns=df.columns.str.strip()
if "Length" in df.columns:
    df['Length']=pd.to_numeric(df['Length'],errors='coerce')
    data=df['Length'].dropna()
    kde=gaussian_kde(data)
    density_values=kde.evaluate(data)
    result=pd.DataFrame({'Length':data,'Density':density_values})
print(result.head())
```

```python
#count by category values  group by manufacyure
manufacture_count=df.groupby('Manufacturer').size().reset_index(name='Count')
print(manufacture_count)
```
